```python
#create empty dataframe
sample_papers = pd.DataFrame(columns = ['ID', 'All content', 'Text', 'Journal_Name', 'class', 'Retracted', 'preprocessed'])
```

```python
#create empty list
samples = []
```

```python
journals_e.shape[0] #amount of journals in total
```

Sample papers from other groups to match largest, per journal: if a journal contains less samples from other groups than non-retracted, oversamples other groups, if not, samples other group

```python
for i in range(0, len(journals_nr)):
  j = journals_nr.iloc[i, 0] #journal name
  a = journals_nr.iloc[i, 1] #number of papers

  if len(train_dataset[(train_dataset['class'] == 'e') & (train_dataset['Journal_Name'] == j)]) < a:
    sample_e = train_dataset[(train_dataset['class'] == 'e') & (train_dataset['Journal_Name'] == j)].sample(a, random_state = 1, replace=True)
  else:
    sample_e = train_dataset[(train_dataset['class'] == 'e') & (train_dataset['Journal_Name'] == j)].sample(a, random_state = 1)
  
  if len(train_dataset[(train_dataset['class'] == 'm') & (train_dataset['Journal_Name'] == j)]) < a:
    sample_m = train_dataset[(train_dataset['class'] == 'm') & (train_dataset['Journal_Name'] == j)].sample(a, random_state = 1, replace=True)
  else:
    sample_m = train_dataset[(train_dataset['class'] == 'm') & (train_dataset['Journal_Name'] == j)].sample(a, random_state = 1)

  sample_nr = train_dataset[(train_dataset['class'] == 'nr') & (train_dataset['Journal_Name'] == j)]
  #appends papers per journal for every group
  samples.append(sample_e)
  samples.append(sample_m)
  samples.append(sample_nr)
```

```python
len(samples) #nested list
```

```python
for i in samples:
  sample_papers = pd.concat([i, sample_papers]) #unnests list
```
