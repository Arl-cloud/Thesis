```python
import pandas as pd
```

```python
df_non_retracted = pd.read_csv('Data (CSV)/non_retracted_papers.csv') 
df_meta_non_retracted = pd.read_csv('Data (CSV)/meta_non_retracted.csv')
```

```python
df_non_retracted['ID'] = df_non_retracted.DOI.str.replace('.pdf', '', regex = True) #remove ".pdf" from ID
```

```python
nr_result = pd.merge(df_non_retracted, df_meta_non_retracted, on="ID") #merge meta and content data
```

```python
nr_result.rename(columns={"Source Title": "Source_Title"}, inplace = True)
```

```python
len(set(nr_result['Source_Title'])) #amount of journals
```

```python
non_retracted_journals = nr_result.groupby(['Source_Title']).size().reset_index(name='Amount').sort_values(by = 'Amount', ascending = False)
non_retracted_journals #check which journals contain how many papers 
```

```python
non_retracted_journals.to_csv('Data (CSV)/non_retracted_journals.csv') #save journal df to csv
```
