```python
!pip install pymupdf
```

```python
import pandas as pd
import fitz  # this is pymupdf
df_misconduct = pd.DataFrame(columns=['DOI', 'Content', 'Retracted'])
for study in files:
  try:
    with fitz.open(study) as doc:
        text = ""
        for page in doc:
            text += page.get_text()
        df_misconduct = df_misconduct.append({'DOI': study, 'Content': text, 'Retracted': 1}, ignore_index=True)
  except:
    pass

df_misconduct
```

```python
df_misconduct.to_csv('/content/drive/MyDrive/Thesis_UU/push/3group/Data (CSV)/retracted_misconduct_papers.csv', index=False, encoding = 'utf-8-sig')
```
