### Creating subsets containing only journals that contain more than one paper in both error and misconduct group

```python
journals_e = e_result.Journal.value_counts() #create df with journals and amount of papers from error group
e_filtered = e_result[e_result.Journal.isin(journals_e.index[journals_e.gt(1)])] #remove all journals which only contain one journal
```

```python
e_filtered.Journal.value_counts().tail(2)
```

```python
journals_e_set = set(e_filtered['Journal'])
print(len(journals_e_set)) #amount of journals
print(journals_e_set) #names of journals
```

```python
sub_misconduct = m_result[m_result.Journal.isin(journals_e_set)] #create subset of misconduct data only with journals from error data
```

```python
len(sub_misconduct)
```

```python
sub_misconduct.head(2)
```

```python
journals_m = sub_misconduct.Journal.value_counts() #create df with journals and amount of papers from misconduct group
sub_misconduct = sub_misconduct[sub_misconduct.Journal.isin(journals_m.index[journals_m.gt(1)])] #remove all journals which only contain one journal
```

```python
sub_misconduct.Journal.value_counts().tail(2)
```

```python
journals_m_set = set(sub_misconduct['Journal'])
print(len(journals_m_set)) #amount of journals
print(journals_m_set) #name of journals
```

```python
sub_error = e_result[e_result.Journal.isin(journals_m_set)] #subsetting error data to only contain journals which appear in misconduct data
```

```python
len(sub_error)
```

```python
sub_error.head(2)
```

```python
sub_retracted = pd.concat([sub_error, sub_misconduct]) #concat to one dataset containing both error and misconduct data
```

```python
len(sub_retracted) #length of created subset containing both groups
```

```python
len(set(sub_retracted['Journal'])) #amount of journals
```

```python
retracted_journals = sub_retracted.groupby(['Journal']).size().reset_index(name='Amount').sort_values(by = 'Amount', ascending = False)
retracted_journals #check which journals contain how many papers 
```

```python
retracted_journals.to_csv('Data (CSV)/retracted_journals.csv') #save journal df to csv
```

```python
sub_retracted.to_csv('Data (CSV)/sub_retracted.csv') #save merged data to csv
```
