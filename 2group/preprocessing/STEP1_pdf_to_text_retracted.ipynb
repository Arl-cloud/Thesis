Scrape text of pdfs, insert into dataframe

```python
!pip install pymupdf
```

```python
import pandas as pd
import fitz  # this is pymupdf
df_retracted = pd.DataFrame(columns=['DOI', 'Content', 'Retracted'])
for study in files:
  try:
    with fitz.open(study) as doc:
        text = ""
        for page in doc:
            text += page.get_text()
        df_retracted = df_retracted.append({'DOI': study, 'Content': text, 'Retracted': 1}, ignore_index=True)
  except:
    pass

df_retracted
```

Save dataframe to a csv file

```python
df_retracted.to_csv('/content/drive/MyDrive/Thesis_UU/push/2group/Data (CSV)/retracted_texts_11_journals.csv', index=False, encoding = 'utf-8-sig')
```
